Q 1: Explain what the forEach method does in JavaScript arrays and write code that uses forEach to display each student's name from an array of student objects to the console.

];
* Ans:-The `forEach()` method in JavaScript arrays executes a provided function once for each array element. It's a way to iterate over an array and perform an action on each item.

Q 2: You have an array of product prices in rupees. Write code using the map method to create a new array that contains the same prices but with 18% GST added to each price.

const originalPrices = [100, 250, 500, 750, 1000];
* Ans:- The `map()` method in JavaScript creates a new array populated with the results of calling a provided function on every element in the calling array.

```python
// Here's the code to add 18% GST to each price:

const originalPrices = [100, 250, 500, 750, 1000];

const pricesWithGST = originalPrices.map(price => {
  const gstAmount = price * 0.18;
  return price + gstAmount;
});

console.log(pricesWithGST);

// The `pricesWithGST` array will contain the following values: [118, 295, 590, 885, 1180].
```

Q 5: Explain how the sort method works in JavaScript and write code to sort an array of employee objects by their salary in descending order (highest salary first).

// Array of employee objects

const employees = [

{ name: "Rajesh", salary: 45000},

{ name: "Priya", salary: 55000},

{ name: "Amit", salary: 40000},

{ name: "Kavya", salary: 60000 }

];
 * Ans:-The `sort()` method in JavaScript is used to sort the elements of an array in place. When sorting numbers, it sorts them in ascending order by default. For more complex sorting, like with objects or specific orders, you provide a comparison function.

This comparison function takes two elements from the array, let's call them `a` and `b`, and determines their order.
*   If it returns a negative value, `a` is placed before `b`.
*   If it returns a positive value, `b` is placed before `a`.
*   If it returns zero, their order remains unchanged relative to each other.

To sort the employee objects by salary in descending order (highest first), you'd want `b.salary - a.salary`. If `b.salary` is larger, the result is positive, putting `b` before `a`.



```python
const employees = [
  { name: "Rajesh", salary: 45000},
  { name: "Priya", salary: 55000},
  { name: "Amit", salary: 40000},
  { name: "Kavya", salary: 60000 }
];

employees.sort((a, b) => b.salary - a.salary);

console.log(employees);


// This will output the array with employees sorted from highest salary to lowest.
```

Q 6: Write code using the reduce method to calculate the total value of all items in a shopping cart. Each item has a price and quantity. Also find the most expensive single item in the cart.

// Shopping cart with items

const cart = [

{ name: "Laptop", price: 50000, quantity: 1},
{ name: "Mouse", price: 500, quantity: 2},
{ name: "Keyboard", price: 1500, quantity: 1},
{ name: "Monitor", price: 15000, quantity: 1}

];
* Ans:-The `reduce()` method in JavaScript executes a reducer function on each element of the array, resulting in a single output value. For calculating the total value, you'll iterate through the cart, multiplying each item's price by its quantity and adding it to an accumulator.

To find the most expensive single item, you can also use `reduce()`. You'll compare the price of the current item with the price of the most expensive item found so far and update if the current item is more expensive.



```python
// Here's the code:
const cart = [
  { name: "Laptop", price: 50000, quantity: 1},
  { name: "Mouse", price: 500, quantity: 2},
  { name: "Keyboard", price: 1500, quantity: 1},
  { name: "Monitor", price: 15000, quantity: 1}
];

// Calculate total value
const totalValue = cart.reduce((accumulator, currentItem) => {
  return accumulator + (currentItem.price * currentItem.quantity);
}, 0); // Start with 0 as the initial accumulator value

// Find the most expensive single item
const mostExpensiveItem = cart.reduce((maxItem, currentItem) => {
  return currentItem.price > maxItem.price ? currentItem : maxItem;
}, cart[0]); // Start with the first item as the initial most expensive

console.log("Total value of items:", totalValue);
console.log("Most expensive single item:", mostExpensiveItem);
```

Q 7: Explain what array destructuring is and write code that extracts the first two colors from an array of colors into separate variables using destructuring.

// Array of colors

const colors ["red", "blue", "green", "yellow", "purple"];
* Ans:-Array destructuring is a JavaScript feature that allows you to unpack values from arrays into distinct variables. It provides a more concise way to access array elements compared to traditional index-based access.


```python
// Here's the code to extract the first two colors using destructuring:
const colors = ["red", "blue", "green", "yellow", "purple"];

const [firstColor, secondColor] = colors;

console.log("First color:", firstColor);
console.log("Second color:", secondColor);
```

* Ans:-

```python
function CalculateTotal(...prices) {
  let total = 0;
  for (const price of prices) {
    total += price;
  }
  return total;
}
```

Q 10: You have two arrays of student names and need to create a new array that contains all students from both arrays. Also, you have two objects representing a student's basic info and marks. Write code using the spread operator to solve both scenarios.

const classA = ["Rahul", "Priya", "Amit"];

const classB ["Sneha", "Vikram", "Pooja"];
* Ans:-
Explanation:

*   For Arrays: The spread operator (`...`) unpacks the elements of `classA` and `classB` individually into the new `allStudents` array. It's like taking all the items out of each array and putting them into a new container.
*   For Objects: Similarly, the spread operator unpacks the properties of `studentInfo` and `studentMarks` into the `completeStudent` object. If there were any duplicate keys, the properties from the later object would overwrite the earlier ones.

```python
// here's how you can use the spread operator for both scenarios:

//Combining Arrays:

const classA = ["Rahul", "Priya", "Amit"];
const classB = ["Sneha", "Vikram", "Pooja"];

const allStudents = [...classA, ...classB];

console.log(allStudents);
// Output: ["Rahul", "Priya", "Amit", "Sneha", "Vikram", "Pooja"]


//Combining Objects (Example):

//Let's say you have these objects:


const studentInfo = { name: "Raj", age: 15 };
const studentMarks = { math: 90, science: 85 };

const completeStudent = { ...studentInfo, ...studentMarks };

console.log(completeStudent);
// Output: { name: "Raj", age: 15, math: 90, science: 85 }
```
